```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop('Cabin', axis=1)
```

```python
print(df.isnull().sum())
```

```python

duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)


df = df.drop_duplicates()
```

```python
#Survival Count
sns.countplot(x='Survived', data=df)
plt.show()
```

```python
#survived by gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.show()
```

```python
#survived by passenger class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.show()
```

```python
#age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.show()
```

```python
#fare distribution
sns.boxplot(x=df['Fare'])
plt.show()
```

```python
#correlation heatmap
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#insights
#1.Females had a higher survival rate than males.
#2.First-class passengers survived more than third-class passengers.
#3.Younger passengers had slightly better survival chances.
#4.Higher ticket fare is related to higher survival probability.
#5.Most passengers were between 20–40 years old.
```
